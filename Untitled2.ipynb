```python
import os
import pandas as pd
import numpy as np
```

```python
DATASET_PATH = 'dataset/'
resp_success = ['지속(정책)', '지속사용(고객혜택제공)', '지속(혜택)']
```

```python
data_path = os.path.join(DATASET_PATH, 'cam_dataset_all.csv')
dataset = pd.read_csv(data_path)
```

```python
def check_YN(resp):
    if resp in resp_success:
        return 'Y'
    else:
        return 'N'
```

```python
dataset['resp_YN']= dataset['반응4'].map(check_YN)
```
